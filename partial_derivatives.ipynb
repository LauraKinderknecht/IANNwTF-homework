**Given multivariate function:**

$
f(x,z,a,b):=y=(4ax^2+a)+3+\sigma(z)+(\sigma(b))^2
$

**Task:**
Calculate the partial derivatives of $f$ w.r.t. each of its variables.

$
\frac{\delta y}{\delta x} 
= \frac{\delta}{\delta x}(4ax^2+a)
+\frac{\delta}{\delta x}3
+\frac{\delta}{\delta x}\sigma(z)
+\frac{\delta}{\delta x}(\sigma(b))^2
\\\\
= 8ax + 0 + 0 + 0
\\\\
= 8ax
$


$
\frac{\delta y}{\delta z} 
= \frac{\delta}{\delta z}(4ax^2+a)
+\frac{\delta}{\delta z}3
+\frac{\delta}{\delta z}\sigma(z)
+\frac{\delta}{\delta z}(\sigma(b))^2
\\\\
= 0 + 0 + \sigma(z) (1 - \sigma(z)) + 0
\\\\
= \sigma(z) (1 - \sigma(z))
$


$
\frac{\delta y}{\delta a} 
= \frac{\delta}{\delta a}(4ax^2+a)
+\frac{\delta}{\delta a}3
+\frac{\delta}{\delta a}\sigma(z)
+\frac{\delta}{\delta a}(\sigma(b))^2
\\\\
= 4x^2+1 + 0 + 0 + 0
\\\\
= 4x^2+1
$

$
\frac{\delta y}{\delta b} 
= \frac{\delta}{\delta b}(4ax^2+a)
+\frac{\delta}{\delta b}3
+\frac{\delta}{\delta b}\sigma(z)
+\frac{\delta}{\delta b}(\sigma(b))^2
\\\\
= 0 + 0 + 0 + \frac{\delta}{\delta b}(\sigma(b))^2
\\\\
= \frac{\delta}{\delta b}(\sigma(b) \cdot \sigma(b))
\\\\
= (\frac{\delta}{\delta b}\sigma(b)) \cdot \sigma(b) + (\frac{\delta}{\delta b}\sigma(b)) \cdot \sigma(b)
\\\\
= 2(\frac{\delta}{\delta b}\sigma(b)) \cdot \sigma(b)
\\\\
= 2 \sigma(b) (1-\sigma(b)) \sigma(b)
\\\\
= 2 \sigma(b)^2 (1-\sigma(b))
$
